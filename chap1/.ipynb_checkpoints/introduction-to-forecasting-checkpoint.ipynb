# Introduction To Forecasting

A **forecast** is a prediction of some future event or events. Most forecasting problems involve the use of time series data. A **time series** is a time-oriented or chronological sequence of observations. A well understood example of a time series is a stock price.

Forecasting problems are classified as
- Short Term (predicting days, weeks, months into the future)
- Medium Term (predicting 1 to 2 years into the future)
- Long Term (can extend many years beyond)

Long-term forecasts impact issues such as strategic planning. Short- and medium-term forecasting is typically based on identifying, modeling, and extrapolating the patterns found in historical data. Because these historical data usually exhibit inertia and do not change dr
amatically very quickly, statistical methods are very useful for short- and medium-term forecasting. We are concerned with these statistical methods.

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('./TGT.csv', parse_dates=['Date'])
df.plot(x='Date', y='Close', figsize=(20, 10))
plt.show() 
```

This is known as a **time series plot**.

There are two broad types of forecasting techniques, **qualitative** methods and **quantitative** methods.


# Qualitative methods

- Are more subjective in nature
- Are used when there is a lack of historical data
- Require the judgement of experts
- Are formalized by the Delphi Method


# Quantitative methods

Quantitative methods make use of historical data and a **forecasting model**. The model formally summarizes patterns in the data and expresses a statistical relationship between previous and current values of the variable. Then the model is used to project the patterns in the data into the future. In other words, the forecasting model is used to extrapolate past and current behavior into the future.

There are several types of forecasting models in general use. The three most widely used are regression models, smoothing models, and general time series models.

```python
df = pd.read_csv('./b4.csv')
df.plot(x='Year', y='Production', figsize=(20, 10), marker='o', title='Pharmaceutical Product Sales', xlabel='Year', ylabel='Production (10^3 lb)')
plt.show()
```

```python
df = pd.read_csv('./b5.csv', parse_dates=['Month'])
df.plot(x='Month', y='Dollars', figsize=(20, 10), marker='o', xlabel='Year', ylabel='Beverage Shipments, Millions of Dollars')
plt.show()
```

The data in this chart exhibits an overall increasing trend, with a distinct **cyclic pattern** that is repeated within each year. The data exhibits monthly or **seasonal** variation.
